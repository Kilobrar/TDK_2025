```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime, timedelta

# List to store the data
data = []

# Base URL
base_url = "https://24.hu/fn/gazdasag/page/"


for i in range(1,660):
    url = f"{base_url}{i}/"
    print(f"Fetching: {url}")
    try:
        # Request the page
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, "html.parser")
        
        # Find all <a> tags with the class "m-articleWidget__link"
        article_links = soup.find_all("a", class_="m-articleWidget__link")
        
        # Extract the href attribute and save it with the page number
        for link in article_links:
            href = link.get("href")  # Get the URL from the href attribute
            if href:  # Ensure href is not None
                data.append({"url": href, "page": i})
    except requests.RequestException as e:
        print(f"Failed to fetch {url}: {e}")


# Save to CSV
df = pd.DataFrame(data)
df.to_csv("24hu_links.csv", index=False)
print("Data saved to 24hu_links.csv")
```

```python
from concurrent.futures import ThreadPoolExecutor

def fetch_article_details(url):
    try:
        print(f"Fetching: {url}")
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, "html.parser")
        
        # Extract the title (from the <h1> tag)
        title_tag = soup.find("h1", class_="o-post__title")
        title = title_tag.get_text(strip=True) if title_tag else "Title not found"
        
        date_part = url.split("gazdasag/")[1][:10]
    
        # Extract the content
        content_div = soup.find("div", class_="o-post__cntWrap")
        if content_div:
            unwanted_classes = ["m-recommendedVideoWidget"]
            for unwanted_class in unwanted_classes:
                for unwanted_div in content_div.find_all("div", class_=unwanted_class):
                    unwanted_div.decompose()  # Remove the unwanted div
                    
            # Get the cleaned content
            content = content_div.get_text(separator=" ", strip=True)
        else:
            content = "Content not found"
            
        return title, content, date_part
    except Exception as e:
        print(f"Error fetching content for {url}: {e}")
        return "Error", "Error fetching content", "Date not found"
    
def process_urls(df, url_column):
    with ThreadPoolExecutor(max_workers=5) as executor:
        results = list(executor.map(fetch_article_details, df[url_column]))
    
    # Unpack results into separate columns
    df["title"], df["content"], df["date"] = zip(*results)
    return df

# Example usage
# Assuming df is your dataframe with a column 'url'
df = process_urls(df, "url")
df.to_csv("24hu_articles.csv", index=False)
print("Data saved to 24hu_articles.csv")
```
