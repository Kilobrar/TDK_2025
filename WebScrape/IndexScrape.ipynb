```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime, timedelta

# Define the date range
start_date = datetime(2024, 12, 31)
end_date = datetime(2018, 1, 1)

# List to store the data
data = []

# Base URL
base_url = "https://index.hu/gazdasag/"

# Loop through the dates
current_date = start_date
while current_date >= end_date:
    date_str = current_date.strftime("%Y/%m/%d")
    url = f"{base_url}{date_str}/"
    print(f"Fetching: {url}")
    try:
        # Request the page
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, "html.parser")
        
        ul = soup.find("ul", id="ajanlok")
        if ul:
            # Gather all <a> tags within the <ul>
            all_links = ul.find_all("a")
            
            # Find all links inside div with class "szerzok_container"
            excluded_links = set()
            for div in soup.find_all("div", class_="szerzok_container"):
                for link in div.find_all("a"):
                    excluded_links.add(link.get("href"))
            
            # Add only links that are not in the excluded_links set
            for link in all_links:
                href = link.get("href")
                if href and href not in excluded_links:
                    data.append({"date": date_str, "url": href})
    except requests.RequestException as e:
        print(f"Failed to fetch {url}: {e}")
    
    # Decrement the date by one day
    current_date -= timedelta(days=1)

# Save to CSV
df = pd.DataFrame(data)
df.to_csv("index_gazdasag_articles.csv", index=False)
print("Data saved to index_gazdasag_articles.csv")
```

```python
from concurrent.futures import ThreadPoolExecutor

def fetch_article_details(url):
    try:
        print(f"Fetching: {url}")
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, "html.parser")
        
        # Extract the title (from the <h1> tag)
        title_tag = soup.find("h1")
        title = title_tag.get_text(strip=True) if title_tag else "Title not found"
    
        # Extract the content
        content_div = soup.find("div", class_="cikk-torzs-container")
        if content_div:
            # Remove unwanted elements
            unwanted_classes = ["cikk-bottom-text-ad", "author-share-date-container", "szelso-jobb"]
            for unwanted_class in unwanted_classes:
                for unwanted_div in content_div.find_all("div", class_=unwanted_class):
                    unwanted_div.decompose()  # Remove the unwanted div
            
            # Get the cleaned content
            content = content_div.get_text(separator=" ", strip=True).replace("\xa0", " ")
        else:
            content = "Content not found"
            
        return title, content
    except Exception as e:
        print(f"Error fetching content for {url}: {e}")
        return "Error", "Error fetching content"
    
def process_urls(df, url_column):
    with ThreadPoolExecutor(max_workers=5) as executor:
        results = list(executor.map(fetch_article_details, df[url_column]))
    
    # Unpack results into separate columns
    df["title"], df["content"] = zip(*results)
    return df

# Example usage
# Assuming df_test is your dataframe with a column 'url'
df = process_urls(df, "url")
df.to_csv("index_gazdasag_articles_with_title_and_cleaned_content.csv", index=False)
print("Data saved to index_gazdasag_articles_with_title_and_cleaned_content.csv")
```
